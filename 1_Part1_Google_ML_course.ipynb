# Imports

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# Define and compile the neural network

```python
# single layer model and single neuron as units = 1 and we have provieded x value as input i.e
# input_shape and our model will predict y value.
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

# optimizer and loss functions are very important, model makes a guess let's say it says y = 5X + 5
# then when training you calculate how bad/good guess is using loss function and then optimizer is used
# to generate another guess and that's how it will get close and close to desire output, it will guess 500 times
# as we have setup epochs = 500.
```

# Providing Data

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

# Train Neural Network

```python
model.fit(xs, ys, epochs=500)
```

# Use Model

```python
print(model.predict([10.0]))
```
